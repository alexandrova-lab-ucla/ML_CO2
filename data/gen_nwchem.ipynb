```python
import os 
```

```python
xyz_files = os.listdir("./xyz/DB3")
print(len(xyz_files))

for i in xyz_files:
    print(i.split(".")[0])
    file_temp = open("./nwchem/" + i.split(".")[0] + ".in", "w+")
    file_temp.write("start " + i.split(".")[0]+"\n")
    file_temp.write("title \"" + i.split(".")[0]+"\"\n")
    file_temp.write("\n")
    file_temp.write("geometry units angstrom\n")
    file_temp.write("  load ./xyz/DB3/" + i + "\n")
    file_temp.write("end\n")
    file_temp.write("basis\n")
    file_temp.write("  * library 3-21g\n")
    file_temp.write("end\n")
    file_temp.write("task scf\n")
    file_temp.write("property vectors " + i.split(".")[0]+".movecs\n")
    file_temp.write("aimfile\n")
    file_temp.write("end\n")
    file_temp.write("task scf property\n")
    file_temp.close()
```
